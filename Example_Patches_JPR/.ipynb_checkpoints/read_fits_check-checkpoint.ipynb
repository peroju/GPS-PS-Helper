```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.io import fits

import pandas as pd
```

```python
def fits_to_arr(fits_file):
    fits_data = fits.open(fits_file)
    im_data = fits_data[0].data.astype(np.float32)
    # print (im_data.shape)
    im_data = np.moveaxis(im_data, 0, -1)
    # print (im_data.shape) # shape here is (512, 512, 3)

    return im_data
```

```python
bkg_im = fits_to_arr('./patch_1_bkg.fits')
src_im = fits_to_arr('./patch_1_counts.fits')

fig = plt.figure(figsize=(8, 5))
fig.add_subplot(121)
plt.imshow(np.sqrt(src_im[:, :, 0]))
plt.colorbar(fraction=0.046, pad=0.04)
fig.add_subplot(122)
plt.imshow(np.sqrt(bkg_im[:, :, 0]))
plt.colorbar(fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()
```

```python
msk_im = fits_to_arr('./patch_1_masks.fits')
msk_im.shape
```

```python
print ('check max min in mask : ', msk_im.max(), msk_im.min())
```

```python
patch_1_csv = pd.read_csv('./patch_1-sources-w_xy.csv')
print (patch_1_csv.shape)
patch_1_csv.head(patch_1_csv.shape[0])
```

```python
x_coords = patch_1_csv['x'].to_numpy()
y_coords = patch_1_csv['y'].to_numpy()
sigma = patch_1_csv['sigma'].to_numpy()
```

```python
import matplotlib.patches as patches

fig = plt.figure(figsize=(10, 6))
fig.add_subplot(121)
plt.imshow(np.sqrt(src_im[:, :, 0]), )
plt.colorbar(fraction=0.046, pad=0.04)
fig.add_subplot(122)
plt.imshow(msk_im[:, :], cmap='gray')


# Plot rectangles around the centers
# sq_w, sq_h = 8, 8

for x, y, s in zip(x_coords, y_coords, sigma):
    # Calculate the coordinates of the top-left corner of the rectangle
    x = x - (s*173.3) / 2
    y = y - (s*173.3) / 2
    # Create a rectangle patch
    rect = patches.Rectangle((x, y), s*173.3, s*173.3, linewidth=1, edgecolor='r', facecolor='none')
    # Add the rectangle patch to the current axes
    plt.gca().add_patch(rect)

plt.title('Masks-P-1-w-BoundingBox')    
plt.colorbar(fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()
```
